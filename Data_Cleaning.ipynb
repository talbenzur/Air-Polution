# Data Cleaning

```python
import pandas as pd
import numpy as np
import requests
```

The final data file contains only one column, so we need to fix it.

```python
data = pd.read_csv (r'result_ex.csv',sep=",")

data.drop("Unnamed: 0",axis='columns', inplace=True)

data=data.rename(columns={'data':'D'})

data = data.D.str.split(r',', expand=True)
data=data.loc[:, 0:9].rename(columns={0:'AQI', 1:'Temperature', 2:'Humidity',3:'Wind',4:'Pressure',5:'Link',6:'City',7:'State',8:'Country'})

data
```

We don't really need Link column, right?

```python
data.drop("Link",axis='columns', inplace=True)
```

Removing empty rows:

```python
def remove_corrupt_rows(df):
    return (df.dropna())
df=remove_corrupt_rows(data)
```

Lets deal with some duplications 

```python
def remove_duplicatives(df, col_name=None):
    return (df.drop_duplicates(subset=col_name, inplace=False))

df=remove_duplicatives(df)
```

Deleting unwanted symbols

```python
for row in df.iterrows():
    for i in range(0,8):
        row[1][i]=''.join(e for e in row[1][i] if e.isalnum())
```
